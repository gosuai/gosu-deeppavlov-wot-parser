# define funcs

```python
#pars p
#pars_p(str(wikicode))
def pars_p(text,page):
    title = ''
    desc = ''
    context = []


    match = re.search('<h2.*?>(.*?)?</h2>', text)
    while match:
        try:
            title = re.search('>(.*?)?</h2>', match.group(0)).group(0)[1:-5]
            
            text = text[match.span(0)[1]:]
            desc_end = re.search('(.*?)?</div>',text).span(0)[1]
            desc = text[:desc_end][:-6]
            desc = re.sub('<.*>','',desc)
            desc = re.sub('{{.*}}','',desc)
            desc = re.sub('\n+','.',desc)
            
            text = text[desc_end:]
            match = re.search('<h2.*?>(.*?)?</h2>', text)
            

            if title != '' and  desc not in ('','.'):
                context.append({
                 'title': title,
                 'desc': desc
                })

                title = ''
                desc = ''
        except:
            text = text[match.span(0)[1]:]
            match = re.search('<h2.*?>(.*?)?</h2>', text)
            continue
    
    result = {'page': page,
              'context': context
             }
    
    return result

       

#pars links
#pars_links(str(wikicode))
def pars_links(text,page):
    links = []
    reg = '\[\[.*?link=(.*?)?(\|+|\]+)'

    match = re.search(reg, text)
    while match:
        if match.group(1) != '':
            links.append(match.group(1))
        text = text[match.span(0)[1]:]
        match = re.search(reg, text)
        
    result = {'page': page,
             'links': links
             }
    
    return result


#pars direct pages
#pars_pages(str(wikicode))
def pars_pages(text,page):
    pages = []
    reg = '\{\{.*?\|.*?\|(.*)?\|.*?\}\}'

    match = re.search(reg, text)
    while match:
        if match.group(1) != '':
            pages.append(match.group(1))
        text = text[match.span(0)[1]:]
        match = re.search(reg, text)
    
    result = {'page': page,
             'pages': pages
             }
    
    return result
```

# load data

```python
site_name = 'wiki.wargaming.net'

meta_page = 'World of Tanks'
```

## tanks context from media wiki

```python
try:
    os.mkdir('./data/tanks/')
except:
    print('already exist')
```

```python
#get meta info of tanks by media wiki
site = mwclient.Site(site_name, path='/')
category = site.categories[meta_page]


#nation tanks
tank_dict = {}

category = site.categories['Tanks']
category = site.categories['Tanks by nation']

for page in category:
    nation = page.name.replace('Category:','')
    tank_dict[nation] = {}
    os.mkdir('./data/tanks/'+nation+'/')
    
    nation_tanks = site.categories[nation]
    for page in nation_tanks:
        tank = page.name.replace('Tank:','')
        wikicode = mwparserfromhell.parse(page.text())
        
        tank_info = str(wikicode)
        tank_info = tank_info.replace('.','. '+tank+' ')
        
        tank_dict[nation][tank] = tank_info
        
        with open('./data/tanks/'+nation+'/'+ tank+ ".txt", "w") as text_file:
            text_file.write(tank_info)
            
        with open('./data/tanks/all tanks/'+ tank+ ".txt", "w") as text_file:
            text_file.write(tank_info)

        
```

```python
import pandas as pd
import cfscrape
from bs4 import BeautifulSoup
import time
import pickle
```

```python
arr = os.listdir('./data/tanks/')
arr = list(set(arr) - set(['all tanks']))

for nation in arr:
    nat_tanks = arr = os.listdir('./data/tanks/' + nation + '/')
    for nat_tank in nat_tanks:
        nat_tank = nat_tank.replace('.txt','')
        context_first = ''

        tank_link = nat_tank.replace(' ','_')
        url = f'https://wiki.wargaming.net/en/Tank:{tank_link}'

        headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

        scraper = cfscrape.create_scraper()

        basket_html = scraper.get(url)

        soup = BeautifulSoup(basket_html.content, 'lxml')

        mydivs = soup.findAll('table')

        for div in mydivs:
            df = pd.read_html(str(div))[0]
            for i , row in df.iterrows():
                for element in row:
                    context_first += str(element) + '. \n'
                    
                    
        file = open('./data/tanks/' + nation + '/'+nat_tank+'.txt')
        mw_context = file.read()
        file.close()
        
        full_context = context_first + mw_context

        
        with open('./data/tanks/'+nation+'/'+ nat_tank+ ".txt", "w") as text_file:
            text_file.write(full_context)
            
        with open('./data/tanks/all tanks/'+ nat_tank+ ".txt", "w") as text_file:
            text_file.write(full_context)
```
